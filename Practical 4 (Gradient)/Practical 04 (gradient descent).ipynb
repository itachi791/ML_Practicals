```python
''' Practical 04: Implement gradient descent algotithm to find the local minima of a function y = (x=3)^2
    starting point = 2
    learning rate = 0.1
    iteration = 20, 30
'''

def f(x):
    return (x + 3) ** 2

def gradientFunction(x):
    return 2 * (x + 3)

def gradientDescent(initial, iteration, learning):

    value = initial
    for i in range(iteration + 1):

        grad = gradientFunction(value)

        value = value - learning * grad

        x_points.append(value)
        y_points.append(((value + 3) **2))

# Driver Code
x_points = list()
y_points = list()
gradientDescent(2, 100, 0.01)

plt.figure(figsize=(10, 6))
plt.plot(x_points, y_points, 'o')
plt.show()
```
