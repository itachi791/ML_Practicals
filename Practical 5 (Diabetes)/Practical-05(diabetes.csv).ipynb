Practical-05: Implement K-Nearest Neighbors algorithm on diabetes.csv dataset. Compute confusion
matrix, accuracy, error rate, precision and recall on the given dataset

dataset = https://www.kaggle.com/datasets/abdallamahgoub/diabetes

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
from google.colab import drive
drive.mount('/content/drive')
```

```python
df = pd.read_csv('/content/drive/MyDrive/BE Practicals/ML/diabetes.csv')
```

```python
df.head()
```

```python
df.isna().sum()
```

```python
y = df.iloc[:, -1]
y
```

```python
X = df.iloc[:, :-1]
X
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=33, random_state=42)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
X_train_scaled = scaler.fit_transform(X_train)
```

```python
X_test_scaled = scaler.transform(X_test)
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
neighbor = KNeighborsClassifier(n_neighbors=7)
```

```python
neighbor.fit(X_train_scaled, y_train)
```

```python
y_pred = neighbor.predict(X_test_scaled)
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score, recall_score, precision_score, f1_score, max_error
```

```python
cm = confusion_matrix(y_test, y_pred)
```

```python
print(cm)
```

```python
accuracy_test = accuracy_score(y_test, y_pred)
accuracy_test
```

```python
precisoin = precision_score(y_test, y_pred)
```

```python
error_rate = max_error(y_test, y_pred)
```

```python
recall = recall_score(y_test, y_pred)
```

```python
y_pred = neighbor.predict(X_train_scaled)
```

```python
accuracy_train = accuracy_score(y_train, y_pred)
```

```python
print('training accuracy =', accuracy_train, 'test accuracy =', accuracy_test, 'Error Rate =', error_rate, 'precisoin= ', precisoin, 'recall =', recall)
```
